```python
import requests
from pprint import pprint as pp
from IPython.display import JSON

country_to_visit = input("What country would you like to visit?: ")
endpoint1 = 'https://restcountries.com/v3.1/name/' + country_to_visit + '?fullText=true' 
print(endpoint1)
response = requests.get(endpoint1)  # making a call to the restcountries.com API
#print(response.status_code)  # check to ensure the api call is successful
data = response.json()
#JSON(data)

for item in data:
    result = dict(capital = item['capital'],
                    continents = item['continents'],
                    currencies = item['currencies'],
                    language = item['languages'],
                    population = item['population'],
                    timezones = item['timezones'])

#print(result)

Intermediate_info = result['currencies']
for key, value in Intermediate_info.items():
        currency_of_countryToVisit = key
#print(currency_of_countryToVisit)

# making a call to the restcountries API
def get_currency(country):
    endpoint = 'https://restcountries.com/v3.1/name/' + country + '?fullText=true' 
    response = requests.get(endpoint)  
    data = response.json()
    #Extracting the right currency data for the curreny of budget
    for item in data:
        result = dict(currencies = item['currencies'])

    #Extract currency abbreviation

    for key, value in result.items():
        currency_dict = value
        for key, value in currency_dict.items():
            currency = key
    return currency


#Getting information from user

budget = float(input("What is your budget \(only numbers\): "))
country = input("which country uses this currency of your budget: ")

#Extract currency abbreviation for the currency the budget is in using the restcountries.com API
currency_of_budget = get_currency(country)

#Obtaining conversion rate using the ExchangeRate-API
api_key = 'd6463bde9e2ea9ae97de9ad2'

endpoint1 = 'https://v6.exchangerate-api.com/v6/d6463bde9e2ea9ae97de9ad2/latest/' + currency_of_budget + '?fullText=true' 

response = requests.get(endpoint1)  # making a call to the the ExchangeRate-API

data = response.json()
# #pp(data)
#Converts the currency you have to us dollars using the ExchangeRate-API
converted_budget = budget * data['conversion_rates'][str(currency_of_countryToVisit)]
print(f"You wish to visit {country_to_visit} for holiday") 
print(f"Your budget is {budget} {currency_of_budget} which is {converted_budget} {currency_of_countryToVisit}.")
print("Enjoy your holidays!!!")
```
